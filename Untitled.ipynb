## **Desafio 1: Leitura dos *chunks***
Neste primeiro desafio, era pedida a leitura de um *chunk* à nossa escolha. Para que a amostra fosse grande e ideal à análise de dados, decidimos que a leitura seria feita a todos os *chunks* do ficheiro. Por esta razão, o seguinte código de execução pode demorar um pouco a correr.

```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")

input_path = 'ghcnd_daily.csv'
output_path = 'ghcnd_daily_processado.csv'

columns_to_keep = ['id', 'year', 'month', 'element'] + [f'value{i}' for i in range(1, 32)]
value_cols = [f'value{i}' for i in range(1, 32)]

# Função de chunking
def process_chunk(chunk):
    chunk.replace(-9999, np.nan, inplace=True)
    chunk = chunk[columns_to_keep]

    chunk['year'] = chunk['year'].astype('int16')
    chunk['month'] = chunk['month'].astype('int8')
    chunk['element'] = chunk['element'].astype('category')
    chunk['id'] = chunk['id'].astype('category')

    # Converter os valores e dividir por 10 para obter °C
    for col in value_cols:
        chunk[col] = chunk[col].astype('float32') / 10.0

    return chunk

# Processar em chunks e guardar no novo ficheiro
chunk_size = 100000
with pd.read_csv(input_path, chunksize=chunk_size) as reader:
    for i, chunk in enumerate(reader):
        processed = process_chunk(chunk)
        processed.to_csv(output_path, mode='a', header=(i == 0), index=False)

print("Processamento completo. Dados guardados em 'ghcnd_daily_processado.csv' com temperaturas em °C.")
```

Este código determina através do **input_path** o ficheiro a ser tratado e para que ficheiro vai a informação processada, através do **output_path**. A seguir, com o **columns_to_keep** e **value_cols**, seleciona as colunas que se pretendem processar. Mais tarde, a função **process_chunk** subtitui valores nulos representados por **-9999** por **Nan**, e otimiza os tipos de dados que queremos processar. Para acerto de valores para a unidade Celsius, nas temperaturas, executamos um **ciclo for** que faz a devida converção dos valores. Depois define-se o processamento em *chunks* com o **chunk_size**, e finalizamos com o armazenamento do ficheiro novo processado e uma mensagem de conclusão da mesma tarefa.

## **Desafio 2: Percentagem de *null values***
Nesta secção, tivemos que calcular a percentagem de valores nulos em cada variável, e determinar qual delas seria a com mais dados em falta. Para tal efetuamos o seguinte código.

```python
import pandas as pd

file_path = 'ghcnd_daily_processado.csv'

# Inicializar contadores
chunk_size = 100000
null_counts = None
total_counts = 0

# Calcular valores nulos em chunks
for chunk in pd.read_csv(file_path, chunksize=chunk_size):
    if null_counts is None:
        null_counts = chunk.isnull().sum()
    else:
        null_counts += chunk.isnull().sum()
    total_counts += len(chunk)

# Calcular percentagem de nulos
percent_nulls = (null_counts / total_counts) * 100
percent_nulls_sorted = percent_nulls.sort_values(ascending=False)

print("\n Percentagem de valores nulos por coluna:")
print(percent_nulls_sorted)

# Mostrar variável com mais dados em falta
worst_col = percent_nulls_sorted.idxmax()
max_null_pct = percent_nulls_sorted.max()
print(f"\n Variável com mais dados em falta: '{worst_col}' ({max_null_pct:.2f}%)")
```

O código analisa valores nulos, através do ficheiro processado anteriormente. No início inicializam-se contadores, tais como o **chunk_size** que define quantas linhas se processam de uma vez, o **null_counts** que conta quantos valores **NaN** existem por coluna, e o **total_counts** que conta o número total de linhas. Após esse passo, é feita contagem de valores nulos por cada *chunk*, com o **chunk.insnull().sum()**, adiciona-se esta contagem ao **null_counts** e atualiza-se o número total de linhas através do **len(chunk)**. Por fim, é feito o cálculo percentual de nulos e é denomida a variável com mais valores em falta.

## Desafio 3: Ano mais antigo e mais recente de registo
Através da sugestão **groupby**, neste desafio o objetivo é determinar para cada estação o ano mais antigo e mais recente de registo de dados. Para tal, definimos 2 dicionários que armazenássem os anos mínimos e máximos, o **min_years** e o **max_years**, respetivamente. A seguir, processamos o ficheiro processado no primeiro desafio em *chunks* com o tamanho definido em **chunk_size** e realizamos um **ciclo for** para que cada *chunk* seja lido nas colunas devidas, e crie novas colunas com os valores armazenados nos dicionários criados, anteriormente. Por fim, criamos um *DataFrame* com os resultados, efetuamos um *reset ao índice para existir a coluna **'id'** e mostramos a amostra dos resultados. Para evitar erros, forçamos a coluna **'year'** a ser numérica.

```python
import pandas as pd

file_path = 'ghcnd_daily_processado.csv'

min_years = {}
max_years = {}

chunk_size = 100000
for chunk in pd.read_csv(file_path, usecols=['id', 'year'], chunksize=chunk_size):
    chunk['year'] = pd.to_numeric(chunk['year'], errors='coerce')  # Ensure 'year' is numeric
    grouped = chunk.groupby('id')['year'].agg(['min', 'max'])

    for station_id, row in grouped.iterrows():
        if station_id in min_years:
            min_years[station_id] = min(min_years[station_id], row['min'])
            max_years[station_id] = max(max_years[station_id], row['max'])
        else:
            min_years[station_id] = row['min']
            max_years[station_id] = row['max']

df_anos = pd.DataFrame({
    'min_year': pd.Series(min_years),
    'max_year': pd.Series(max_years)
})

df_anos.reset_index(inplace=True)
df_anos.rename(columns={'index': 'station_id'}, inplace=True)

print("\n Intervalo de anos por estação:")
print(df_anos.head())
```

## **Desafio 4: Determinar temperatura média**
Para realizar esta fase, foi necessário definir um ficheiro de entrada **'ghcnd_daily_processado.csv'** e de saída **'ghcnd_daily_com_media.csv'**. DEpois definiu-se a lista de colunas de temperatura diária com o **value_cols**. Após processar em *chunks*, fazemos um **ciclo for** que lê o ficheiro de entrada e calcula a média das temperaturas diárias por linha, ignorando os valores NaN. Estes valores vão ser guardados como uma nova coluna, e quando esta operação acaba, aparece uma notificação. Para evitar erros forçamos os valores das temperaturas a serem lidos como **float**.

```python
import pandas as pd

file_path = 'ghcnd_daily_processado.csv'
output_path = 'ghcnd_daily_com_media.csv'

# Lista de colunas de temperatura diárias
value_cols = [f'value{i}' for i in range(1, 32)]

chunk_size = 100000
for i, chunk in enumerate(pd.read_csv(file_path, chunksize=chunk_size)):
    # Garantir que os valores das colunas de temperatura sejam float32
    for col in value_cols:
        chunk[col] = pd.to_numeric(chunk[col], errors='coerce').astype('float32')

    # Calcular a média das temperaturas diárias por linha (ignorando NaNs)
    chunk['daily_avg_temp'] = chunk[value_cols].mean(axis=1)

    # Mostrar as primeiras 5 linhas do primeiro chunk
    if i == 0:
        print(chunk.head())

    # Guardar com a nova coluna
    chunk.to_csv(output_path, mode='a', header=(i == 0), index=False)

print("Temperatura média diária (em °C) calculada e guardada em 'ghcnd_daily_com_media.csv'.")
```

## **Desafio 5: Agrupar dados por nome de estação e ano utilizando a função mean**

## **Desafio 6: Selecionar as estações portuguesas**
Neste exercício começamos por ler os dados do *ghcnd-stations.txt* e filtra pelas estações portuguesas com nomes específicos e tbm com IDs que começam por PO e POM, após obter esses ids ele vai ler os dados do *ghcnd_daily_processado.csv* só mantendo colunas relevantes, após a filtragem ele vai dar append desses dados para um novo ficheiro csv chamado "estacoesPT.csv

```python
import pandas as pd

dados_path = 'ghcnd_daily_processado.csv'
stations_path = 'global_climate_data/ghcnd-stations.txt'
output_path = 'estacoesPT.csv'

# Estações a procurar (nomes)
estacoes_pt_nomes = ['HORTA', 'FUNCHAL', 'LISBOA', 'CASTELO BRANCO', 'FARO']

stations_cols = ['id', 'latitude', 'longitude', 'elevation', 'state', 'name']
stations = pd.read_fwf(stations_path, header=None, widths=[11, 9, 10, 7, 3, 31], names=stations_cols)

# Filtrar pelas estações portuguesas (nome + prefixo do país)
stations_pt = stations[
    stations['name'].str.contains('|'.join(estacoes_pt_nomes), case=False, na=False) &
    stations['id'].str.startswith(('PO', 'POM'))
]


# Obter os IDs das estações filtradas
ids_pt = stations_pt['id'].tolist()

# Colunas a manter
colunas = ['id', 'year', 'month', 'element'] + [f'value{i}' for i in range(1, 32)]

# Filtrar os dados por ID
chunk_size = 100000
for i, chunk in enumerate(pd.read_csv(dados_path, usecols=colunas, chunksize=chunk_size)):
    chunk_filtrado = chunk[chunk['id'].isin(ids_pt)]
    chunk_filtrado.to_csv(output_path, mode='a', header=(i == 0), index=False)

print(" Dados das estações portuguesas extraídos para 'estacoesPT.csv'.")

df = pd.read_csv(output_path)
print(df)
```

## **Desafio 7: Substituição de ids por nomes**

- **Desafio extra - 1**: Gráfico de comparação de memória antes e depois do processamento.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

input_path = 'ghcnd_daily.csv'
columns_to_keep = ['id', 'year', 'month', 'element'] + [f'value{i}' for i in range(1, 32)]
value_cols = [f'value{i}' for i in range(1, 32)]

# Especificar dtype para evitar warnings
dtype_spec = {
    'id': 'object', 'year': 'int32', 'month': 'int32', 'element': 'object',
    **{f'value{i}': 'float32' for i in range(1, 32)}
}

# Ler e limpar chunk
chunk = pd.read_csv(input_path, nrows=100000, dtype=dtype_spec, na_values=-9999)
chunk = chunk[columns_to_keep]

# Uso de memória original
mem_before = chunk.memory_usage(deep=True) / 1024  # KB

# Processar
chunk['year'] = chunk['year'].astype('int16')
chunk['month'] = chunk['month'].astype('int8')
chunk['element'] = chunk['element'].astype('category')
chunk['id'] = chunk['id'].astype('category')
for col in value_cols:
    chunk[col] = chunk[col].astype('float32')

# Uso de memória após
mem_after = chunk.memory_usage(deep=True) / 1024

# Comparação
mem_comparison = pd.DataFrame({'Antes (KB)': mem_before, 'Depois (KB)': mem_after})
mem_comparison = mem_comparison.loc[columns_to_keep]

# Plot
mem_comparison.plot(kind='bar', figsize=(14, 6), color=['red', 'green'])
plt.title('Uso de Memória por Coluna Antes vs Depois da Conversão de Tipos')
plt.ylabel('Memória (KB)')
plt.xlabel('Colunas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(loc='upper right')
plt.show()
```

- **Desafio extra - 2**: Percentagem de valores nulos

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

file_path = 'ghcnd_daily_processado.csv'

chunk_size = 100000
null_counts = None
total_counts = 0

# Calcular valores nulos em chunks
for chunk in pd.read_csv(file_path, chunksize=chunk_size):
    if null_counts is None:
        null_counts = chunk.isnull().sum()
    else:
        null_counts += chunk.isnull().sum()
    total_counts += len(chunk)

# Calcular percentagem de nulos
percent_nulls = (null_counts / total_counts) * 100
percent_nulls_sorted = percent_nulls.sort_values(ascending=False)

# Mostrar resultados
print("\n Percentagem de valores nulos por coluna:")
print(percent_nulls_sorted)

# Mostrar variável com mais dados em falta
worst_col = percent_nulls_sorted.idxmax()
max_null_pct = percent_nulls_sorted.max()
print(f"\n Variável com mais dados em falta: '{worst_col}' ({max_null_pct:.2f}%)")

# Criar gráfico de barras
plt.figure(figsize=(12, 6))
percent_nulls_sorted.plot(kind='bar', color='skyblue')
plt.title('Percentagem de Valores Nulos por Coluna')
plt.ylabel('Percentagem (%)')
plt.xlabel('Colunas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

- **Desafio extra - 3**: Mapa das estações portuguesas

```python
import pandas as pd
import folium

stations_path = 'global_climate_data/ghcnd-stations.txt'
estacoes_pt_nomes = ['HORTA', 'FUNCHAL', 'LISBOA', 'CASTELO BRANCO', 'FARO']
stations_cols = ['id', 'latitude', 'longitude', 'elevation', 'state', 'name']

stations = pd.read_fwf(stations_path, header=None, widths=[11, 9, 10, 7, 3, 31], names=stations_cols)

# Filtrar por nome e por prefixo de país (PO ou POM → Portugal)
stations_pt = stations[
    stations['name'].str.contains('|'.join(estacoes_pt_nomes), case=False) &
    stations['id'].str.startswith(('PO', 'POM'))
]

# Mostrar resultado no terminal
print(" Estações meteorológicas portuguesas encontradas:")
print(stations_pt[['id', 'name', 'latitude', 'longitude']].to_string(index=False))

# Criar mapa centrado em Portugal
mapa = folium.Map(location=[39.5, -8.0], zoom_start=6)

# Adicionar marcadores
for _, row in stations_pt.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"{row['name']} ({row['id']})",
        tooltip=row['name'],
        icon=folium.Icon(color='blue', icon='cloud')
    ).add_to(mapa)

mapa.save("estacoes_portuguesas.html")
print(" Mapa gerado como 'estacoes_portuguesas.html'")
```
