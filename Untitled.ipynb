## **Desafio 1: Leitura dos *chunks***
Neste primeiro desafio, era pedida a leitura de um *chunk* √† nossa escolha. Para que a amostra fosse grande e ideal √† an√°lise de dados, decidimos que a leitura seria feita a todos os *chunks* do ficheiro. Por esta raz√£o, o seguinte c√≥digo de execu√ß√£o pode demorar um pouco a correr.

```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")

input_path = 'ghcnd_daily.csv'
output_path = 'ghcnd_daily_processado.csv'

columns_to_keep = ['id', 'year', 'month', 'element'] + [f'value{i}' for i in range(1, 32)]
value_cols = [f'value{i}' for i in range(1, 32)]

# Fun√ß√£o de chunking
def process_chunk(chunk):
    chunk.replace(-9999, np.nan, inplace=True)
    chunk = chunk[columns_to_keep]

    chunk['year'] = chunk['year'].astype('int16')
    chunk['month'] = chunk['month'].astype('int8')
    chunk['element'] = chunk['element'].astype('category')
    chunk['id'] = chunk['id'].astype('category')

    # Converter os valores e dividir por 10 para obter ¬∞C
    for col in value_cols:
        chunk[col] = chunk[col].astype('float32') / 10.0

    return chunk

# Processar em chunks e guardar no novo ficheiro
chunk_size = 100000
with pd.read_csv(input_path, chunksize=chunk_size) as reader:
    for i, chunk in enumerate(reader):
        processed = process_chunk(chunk)
        processed.to_csv(output_path, mode='a', header=(i == 0), index=False)

print("Processamento completo. Dados guardados em 'ghcnd_daily_processado.csv' com temperaturas em ¬∞C.")
```

Este c√≥digo determina atrav√©s do **input_path** o ficheiro a ser tratado e para que ficheiro vai a informa√ß√£o processada, atrav√©s do **output_path**. A seguir, com o **columns_to_keep** e **value_cols**, seleciona as colunas que se pretendem processar. Mais tarde, a fun√ß√£o **process_chunk** subtitui valores nulos representados por **-9999** por **Nan**, e otimiza os tipos de dados que queremos processar. Para acerto de valores para a unidade Celsius, nas temperaturas, executamos um **ciclo for** que faz a devida conver√ß√£o dos valores. Depois define-se o processamento em *chunks* com o **chunk_size**, e finalizamos com o armazenamento do ficheiro novo processado e uma mensagem de conclus√£o da mesma tarefa.

## **Desafio 2: Percentagem de *null values***
Nesta sec√ß√£o, tivemos que calcular a percentagem de valores nulos em cada vari√°vel, e determinar qual delas seria a com mais dados em falta. Para tal efetuamos o seguinte c√≥digo.

```python
import pandas as pd

file_path = 'ghcnd_daily_processado.csv'

# Inicializar contadores
chunk_size = 100000
null_counts = None
total_counts = 0

# Calcular valores nulos em chunks
for chunk in pd.read_csv(file_path, chunksize=chunk_size):
    if null_counts is None:
        null_counts = chunk.isnull().sum()
    else:
        null_counts += chunk.isnull().sum()
    total_counts += len(chunk)

# Calcular percentagem de nulos
percent_nulls = (null_counts / total_counts) * 100
percent_nulls_sorted = percent_nulls.sort_values(ascending=False)

print("\n Percentagem de valores nulos por coluna:")
print(percent_nulls_sorted)

# Mostrar vari√°vel com mais dados em falta
worst_col = percent_nulls_sorted.idxmax()
max_null_pct = percent_nulls_sorted.max()
print(f"\n Vari√°vel com mais dados em falta: '{worst_col}' ({max_null_pct:.2f}%)")
```

O c√≥digo analisa valores nulos, atrav√©s do ficheiro processado anteriormente. No in√≠cio inicializam-se contadores, tais como o **chunk_size** que define quantas linhas se processam de uma vez, o **null_counts** que conta quantos valores **NaN** existem por coluna, e o **total_counts** que conta o n√∫mero total de linhas. Ap√≥s esse passo, √© feita contagem de valores nulos por cada *chunk*, com o **chunk.insnull().sum()**, adiciona-se esta contagem ao **null_counts** e atualiza-se o n√∫mero total de linhas atrav√©s do **len(chunk)**. Por fim, √© feito o c√°lculo percentual de nulos e √© denomida a vari√°vel com mais valores em falta.

## Desafio 3: Ano mais antigo e mais recente de registo
Atrav√©s da sugest√£o **groupby**, neste desafio o objetivo √© determinar para cada esta√ß√£o o ano mais antigo e mais recente de registo de dados. Para tal, definimos 2 dicion√°rios que armazen√°ssem os anos m√≠nimos e m√°ximos, o **min_years** e o **max_years**, respetivamente. A seguir, processamos o ficheiro processado no primeiro desafio em *chunks* com o tamanho definido em **chunk_size** e realizamos um **ciclo for** para que cada *chunk* seja lido nas colunas devidas, e crie novas colunas com os valores armazenados nos dicion√°rios criados, anteriormente. Por fim, criamos um *DataFrame* com os resultados, efetuamos um *reset ao √≠ndice para existir a coluna **'id'** e mostramos a amostra dos resultados. Para evitar erros, for√ßamos a coluna **'year'** a ser num√©rica.

```python
import pandas as pd

file_path = 'ghcnd_daily_processado.csv'

min_years = {}
max_years = {}

chunk_size = 100000
for chunk in pd.read_csv(file_path, usecols=['id', 'year'], chunksize=chunk_size):
    chunk['year'] = pd.to_numeric(chunk['year'], errors='coerce')  # Ensure 'year' is numeric
    grouped = chunk.groupby('id')['year'].agg(['min', 'max'])

    for station_id, row in grouped.iterrows():
        if station_id in min_years:
            min_years[station_id] = min(min_years[station_id], row['min'])
            max_years[station_id] = max(max_years[station_id], row['max'])
        else:
            min_years[station_id] = row['min']
            max_years[station_id] = row['max']

df_anos = pd.DataFrame({
    'min_year': pd.Series(min_years),
    'max_year': pd.Series(max_years)
})

df_anos.reset_index(inplace=True)
df_anos.rename(columns={'index': 'station_id'}, inplace=True)

print("\n Intervalo de anos por esta√ß√£o:")
print(df_anos.head())
```

## **Desafio 4: Determinar temperatura m√©dia**
Para realizar esta fase, foi necess√°rio definir um ficheiro de entrada **'ghcnd_daily_processado.csv'** e de sa√≠da **'ghcnd_daily_com_media.csv'**. DEpois definiu-se a lista de colunas de temperatura di√°ria com o **value_cols**. Ap√≥s processar em *chunks*, fazemos um **ciclo for** que l√™ o ficheiro de entrada e calcula a m√©dia das temperaturas di√°rias por linha, ignorando os valores NaN. Estes valores v√£o ser guardados como uma nova coluna, e quando esta opera√ß√£o acaba, aparece uma notifica√ß√£o. Para evitar erros for√ßamos os valores das temperaturas a serem lidos como **float**.

```python
import pandas as pd

file_path = 'ghcnd_daily_processado.csv'
output_path = 'ghcnd_daily_com_media.csv'

# Lista de colunas de temperatura di√°rias
value_cols = [f'value{i}' for i in range(1, 32)]

chunk_size = 100000
for i, chunk in enumerate(pd.read_csv(file_path, chunksize=chunk_size)):
    # Garantir que os valores das colunas de temperatura sejam float32
    for col in value_cols:
        chunk[col] = pd.to_numeric(chunk[col], errors='coerce').astype('float32')

    # Calcular a m√©dia das temperaturas di√°rias por linha (ignorando NaNs)
    chunk['daily_avg_temp'] = chunk[value_cols].mean(axis=1)

    # Mostrar as primeiras 5 linhas do primeiro chunk
    if i == 0:
        print(chunk.head())

    # Guardar com a nova coluna
    chunk.to_csv(output_path, mode='a', header=(i == 0), index=False)

print("Temperatura m√©dia di√°ria (em ¬∞C) calculada e guardada em 'ghcnd_daily_com_media.csv'.")
```

## **Desafio 5: Agrupar dados por nome de esta√ß√£o e ano utilizando a fun√ß√£o mean**


```python
import pandas as pd
import numpy as np

# Caminhos dos ficheiros
file_path = 'ghcnd_daily_processado.csv'
stations_path = 'global_climate_data/ghcnd-stations.txt'
output_path = 'ghcnd_daily_com_media.csv'

# Colunas de valores di√°rios
value_cols = [f'value{i}' for i in range(1, 32)]

# Tamanho do chunk
chunk_size = 100000

# Processar em chunks e calcular m√©dia di√°ria
for i, chunk in enumerate(pd.read_csv(file_path, chunksize=chunk_size)):
    # Substituir -9999 por NaN
    chunk[value_cols] = chunk[value_cols].replace(-9999, np.nan)
    
    # Calcular m√©dia das temperaturas por linha
    chunk['daily_avg_temp'] = chunk[value_cols].mean(axis=1)
    
    # Salvar para ficheiro com nova coluna
    chunk.to_csv(output_path, mode='a', header=(i == 0), index=False)

print(" Temperatura m√©dia di√°ria calculada e guardada em 'ghcnd_daily_com_media.csv'.")

# ---------------------------------------------------------------
# 2. Agrupar por nome da esta√ß√£o e ano

# Ler dados com m√©dia calculada
df = pd.read_csv(output_path, usecols=['id', 'year', 'daily_avg_temp'])

# Ler nome das esta√ß√µes
stations = pd.read_fwf(stations_path, header=None, names=['id', 'latitude', 'longitude', 'elevation', 'state', 'name'],
                       widths=[11, 9, 10, 7, 3, 30], dtype=str)

# Juntar nome da esta√ß√£o aos dados
df = df.merge(stations[['id', 'name']], on='id', how='left')

# Agrupar por nome e ano e calcular a m√©dia da temperatura di√°ria
grouped = df.groupby(['name', 'year'])['daily_avg_temp'].mean().reset_index()

# Mostrar primeiros resultados
print("\n M√©dia de temperatura di√°ria por esta√ß√£o e ano:")
print(grouped.head())

# (Opcional) Guardar em CSV
grouped.to_csv('media_por_estacao_ano.csv', index=False)
print("\n Resultado guardado em 'media_por_estacao_ano.csv'")
```

## **Desafio 6: Selecionar as esta√ß√µes portuguesas**
Neste exerc√≠cio come√ßamos por ler os dados do *ghcnd-stations.txt* e filtra pelas esta√ß√µes portuguesas com nomes espec√≠ficos e tbm com IDs que come√ßam por PO e POM, ap√≥s obter esses ids ele vai ler os dados do *ghcnd_daily_processado.csv* s√≥ mantendo colunas relevantes, ap√≥s a filtragem ele vai dar append desses dados para um novo ficheiro csv chamado "estacoesPT.csv

```python
import pandas as pd

dados_path = 'ghcnd_daily_processado.csv'
stations_path = 'global_climate_data/ghcnd-stations.txt'
output_path = 'estacoesPT.csv'

# Esta√ß√µes a procurar (nomes)
estacoes_pt_nomes = ['HORTA', 'FUNCHAL', 'LISBOA', 'CASTELO BRANCO', 'FARO']

stations_cols = ['id', 'latitude', 'longitude', 'elevation', 'state', 'name']
stations = pd.read_fwf(stations_path, header=None, widths=[11, 9, 10, 7, 3, 31], names=stations_cols)

# Filtrar pelas esta√ß√µes portuguesas (nome + prefixo do pa√≠s)
stations_pt = stations[
    stations['name'].str.contains('|'.join(estacoes_pt_nomes), case=False, na=False) &
    stations['id'].str.startswith(('PO', 'POM'))
]


# Obter os IDs das esta√ß√µes filtradas
ids_pt = stations_pt['id'].tolist()

# Colunas a manter
colunas = ['id', 'year', 'month', 'element'] + [f'value{i}' for i in range(1, 32)]

# Filtrar os dados por ID
chunk_size = 100000
for i, chunk in enumerate(pd.read_csv(dados_path, usecols=colunas, chunksize=chunk_size)):
    chunk_filtrado = chunk[chunk['id'].isin(ids_pt)]
    chunk_filtrado.to_csv(output_path, mode='a', header=(i == 0), index=False)

print(" Dados das esta√ß√µes portuguesas extra√≠dos para 'estacoesPT.csv'.")

df = pd.read_csv(output_path)
print(df)
```

## **Desafio 7: Substitui√ß√£o de ids por nomes**
Neste desafio, o c√≥digo ao ler o ficheiro estacoesPT.csv consegue obter o id das esta√ß√µes portuguesas e comparando com o ficheiro *ghcnd-stations.txt* ele s√≥ obt√©m os IDs e nomes e d√° merge dos dados obtidos para substituir os ids com o seu nome correspondente e d√° rename da coluna "id" para "station_name" e salva para o ficheiro id_estacoes.csv

```python
import pandas as pd

dados_path = 'estacoesPT.csv'
stations_path = 'global_climate_data/ghcnd-stations.txt'
output_path = 'id_estacoesPT.csv'

# Ler ficheiro de esta√ß√µes
stations_cols = ['id', 'latitude', 'longitude', 'elevation', 'state', 'name']
stations = pd.read_fwf(stations_path, header=None, widths=[11, 9, 10, 7, 3, 31], names=stations_cols)
stations = stations[['id', 'name']].drop_duplicates()

# Carregar os dados filtrados e substituir IDs por nomes
df = pd.read_csv(dados_path)
df = df.merge(stations, on='id', how='left')
df.drop(columns=['id'], inplace=True)
df.rename(columns={'name': 'station_name'}, inplace=True)

df.to_csv(output_path, index=False)

print(" IDs substitu√≠dos por nomes das esta√ß√µes em 'id_estacoesPT.csv'.")
```

O c√≥digo come√ßa por ler o ficheiro *ghcnd-station.txt* que identifica as esta√ß√µes existentes e a partir da√≠ filtramos pelo nome "Castelo Branco", caso uma esta√ß√£o seja encontrada ele d√° print das informa√ß√µes dessa esta√ß√£o. De seguida usamos o Folium tal como recomendado pela professor para a cria√ß√£o do mapa onde adicionamos o marcador da localiza√ß√£o, ap√≥s isso o ficheiro HTML √© guardado com o nome "mapa.html".

```python
import pandas as pd
import folium

stations_path = 'global_climate_data/ghcnd-stations.txt'
keywords_pt = ['CASTELO BRANCO']

# Ler ficheiro de esta√ß√µes
stations_cols = ['id', 'latitude', 'longitude', 'elevation', 'state', 'name']
stations = pd.read_fwf(
    stations_path, header=None,
    widths=[11, 9, 10, 7, 3, 31],
    names=stations_cols
)

# Filtrar apenas CASTELO BRANCO
stations['name_clean'] = stations['name'].str.upper()
mask = stations['name_clean'].str.contains('|'.join(keywords_pt), case=False, regex=True)
stations_cb = stations[mask].copy()

# Verificar se encontrou
print("\n Esta√ß√£o encontrada:")
print(stations_cb[['id', 'name', 'latitude', 'longitude']])

# Criar o mapa centrado em Castelo Branco (ou usar a localiza√ß√£o da esta√ß√£o)
if not stations_cb.empty:
    cb_lat = stations_cb.iloc[0]['latitude']
    cb_lon = stations_cb.iloc[0]['longitude']
    m = folium.Map(location=[cb_lat, cb_lon], zoom_start=10)

    # Adicionar marcador
    folium.Marker(
        location=[cb_lat, cb_lon],
        popup=stations_cb.iloc[0]['name'].strip(),
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(m)

    # Guardar como HTML
    m.save('mapa.html')
    print(" Mapa criado: 'mapa.html'")
else:
    print(" Nenhuma esta√ß√£o encontrada com o nome 'CASTELO BRANCO'.")
```

- **Desafio extra - 1**: Gr√°fico de compara√ß√£o de mem√≥ria antes e depois do processamento.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

input_path = 'ghcnd_daily.csv'
columns_to_keep = ['id', 'year', 'month', 'element'] + [f'value{i}' for i in range(1, 32)]
value_cols = [f'value{i}' for i in range(1, 32)]

# Especificar dtype para evitar warnings
dtype_spec = {
    'id': 'object', 'year': 'int32', 'month': 'int32', 'element': 'object',
    **{f'value{i}': 'float32' for i in range(1, 32)}
}

# Ler e limpar chunk
chunk = pd.read_csv(input_path, nrows=100000, dtype=dtype_spec, na_values=-9999)
chunk = chunk[columns_to_keep]

# Uso de mem√≥ria original
mem_before = chunk.memory_usage(deep=True) / 1024  # KB

# Processar
chunk['year'] = chunk['year'].astype('int16')
chunk['month'] = chunk['month'].astype('int8')
chunk['element'] = chunk['element'].astype('category')
chunk['id'] = chunk['id'].astype('category')
for col in value_cols:
    chunk[col] = chunk[col].astype('float32')

# Uso de mem√≥ria ap√≥s
mem_after = chunk.memory_usage(deep=True) / 1024

# Compara√ß√£o
mem_comparison = pd.DataFrame({'Antes (KB)': mem_before, 'Depois (KB)': mem_after})
mem_comparison = mem_comparison.loc[columns_to_keep]

# Plot
mem_comparison.plot(kind='bar', figsize=(14, 6), color=['red', 'green'])
plt.title('Uso de Mem√≥ria por Coluna Antes vs Depois da Convers√£o de Tipos')
plt.ylabel('Mem√≥ria (KB)')
plt.xlabel('Colunas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(loc='upper right')
plt.show()
```

- **Desafio extra - 2**: Percentagem de valores nulos

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

file_path = 'ghcnd_daily_processado.csv'

chunk_size = 100000
null_counts = None
total_counts = 0

# Calcular valores nulos em chunks
for chunk in pd.read_csv(file_path, chunksize=chunk_size):
    if null_counts is None:
        null_counts = chunk.isnull().sum()
    else:
        null_counts += chunk.isnull().sum()
    total_counts += len(chunk)

# Calcular percentagem de nulos
percent_nulls = (null_counts / total_counts) * 100
percent_nulls_sorted = percent_nulls.sort_values(ascending=False)

# Mostrar resultados
print("\n Percentagem de valores nulos por coluna:")
print(percent_nulls_sorted)

# Mostrar vari√°vel com mais dados em falta
worst_col = percent_nulls_sorted.idxmax()
max_null_pct = percent_nulls_sorted.max()
print(f"\n Vari√°vel com mais dados em falta: '{worst_col}' ({max_null_pct:.2f}%)")

# Criar gr√°fico de barras
plt.figure(figsize=(12, 6))
percent_nulls_sorted.plot(kind='bar', color='skyblue')
plt.title('Percentagem de Valores Nulos por Coluna')
plt.ylabel('Percentagem (%)')
plt.xlabel('Colunas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

- **Desafio extra - 3**: Mapa das esta√ß√µes portuguesas

```python
import pandas as pd
import folium

stations_path = 'global_climate_data/ghcnd-stations.txt'
estacoes_pt_nomes = ['HORTA', 'FUNCHAL', 'LISBOA', 'CASTELO BRANCO', 'FARO']
stations_cols = ['id', 'latitude', 'longitude', 'elevation', 'state', 'name']

stations = pd.read_fwf(stations_path, header=None, widths=[11, 9, 10, 7, 3, 31], names=stations_cols)

# Filtrar por nome e por prefixo de pa√≠s (PO ou POM ‚Üí Portugal)
stations_pt = stations[
    stations['name'].str.contains('|'.join(estacoes_pt_nomes), case=False) &
    stations['id'].str.startswith(('PO', 'POM'))
]

# Mostrar resultado no terminal
print(" Esta√ß√µes meteorol√≥gicas portuguesas encontradas:")
print(stations_pt[['id', 'name', 'latitude', 'longitude']].to_string(index=False))

# Criar mapa centrado em Portugal
mapa = folium.Map(location=[39.5, -8.0], zoom_start=6)

# Adicionar marcadores
for _, row in stations_pt.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"{row['name']} ({row['id']})",
        tooltip=row['name'],
        icon=folium.Icon(color='blue', icon='cloud')
    ).add_to(mapa)

mapa.save("estacoes_portuguesas.html")
print(" Mapa gerado como 'estacoes_portuguesas.html'")
```
